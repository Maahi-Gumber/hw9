**Description of the problem**

Credit card defaults represent a significant financial risk for banks and lending institutions, with billions of dollars in losses annually . This analysis explores how machine learning can help predict which customers are likely to default on their credit card payments, potentially enabling proactive intervention and risk mitigation strategie

**Description of Dataset**

The dataset contains comprehensive information about 21,000 credit card customers with 24 features spanning various categories of information . The target variable, default.payment.next.month, indicates whether a customer defaulted (1) or not (0) in the following month.



**Description of the model**
I tried three progressively more sophisticated pipelines, each wrapped in sklearn’s Pipeline/ColumnTransformer, with one-hot encoding for categoricals and standard scaling for numerics.

| Model          | Train Score    | Test Score     | Fit Time (s)    | Score Time (s)   |
|:--------------:|:-------------:|:--------------:|:---------------:|:----------------:|
| Decision Tree  | 0.999 ± 0.000 | 0.697 ± 0.003  | 0.673 ± 0.025   | 0.024 ± 0.001    | 
| kNN            | 0.811 ± 0.001 | 0.737 ± 0.005  | 0.063 ± 0.002   | 0.565 ± 0.017    | 
| RBF SVM        | 0.778 ± 0.000 | 0.777 ± 0.000  | 25.917 ± 0.627  | 9.269 ± 0.252    | 

The tree dramatically over-fitted; kNN yielded an acceptable baseline; the RBF-SVM gave the best generalisation in spite of a lower training score.

**Feature Engineering**

To enhance model performance, I created two engineered features . The UTILIZATION feature represents the ratio of current bill amount to credit limit, providing insight into credit usage patterns . The AGE_GROUP feature consists of categorical age buckets (20s, 30s, 40s, 50s) to capture non-linear age effects .

The preprocessing pipeline included standard scaling for numerical features, one-hot encoding for categorical variables, and careful handling of missing values . For demographic features with inconsistent encoding, I used constant imputation with a "missing" category to preserve information about data quality issues .

I systematically evaluated four different machine learning algorithms: Logistic Regression, Decision Trees, k-Nearest Neighbors (kNN), and Support Vector Machines with RBF kernel . Each model was optimized through cross-validation and hyperparameter tuning

**Results**

RBF SVM and optimized Logistic Regression achieved the highest test accuracy at 77.7%, while showing good generalization with minimal overfitting . Decision Trees without depth constraints severely overfit (99.9% training accuracy vs. 69.7% test accuracy), demonstrating the importance of regularization . The kNN algorithm showed moderate performance with reasonable generalization capabilities.



**Caveats**

*Temporal Data Leakage:* The dataset represents a specific time period in Taiwan's credit market . Economic conditions, regulatory changes, or cultural shifts in payment behavior could significantly impact model performance when applied to future data or different markets.

*Demographic and Cultural Specificity:* The dataset originates from Taiwan and may reflect culture-specific financial behaviors, regulatory environments, and economic conditions that don't generalize to other markets . Payment patterns, family financial support systems, and social attitudes toward debt vary significantly across cultures, potentially limiting the model's applicability to diverse customer populations

**Conclusions**

This project shows that machine learning can do a good job predicting when people might default on their credit cards, using the information already available about customers. It performed better than basic methods, especially when using models like RBF SVM and logistic regression that were set up carefully to avoid overfitting.

But to actually use these models in the real world, we need to think about things like how stable the data is over time, cultural differences, and how the business operates.

Overall, the project shows that using data to assess financial risk has a lot of potential—but it also has its limits. It's important to not just rely on the numbers, but to also understand the industry and keep testing and improving the models as things change.
